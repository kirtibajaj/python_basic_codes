```python
# substraction (**only work at numbers datatype**)
#Associativity of operation(+,-,*,/) is left to right
x=3
y=7
z=x-y
print(x)
print(y)
print(z)
```

```python
#Multiplication

x=1
y=2.4
z=3.14+4.5j
r=x*y*z
print(r)
```

```python
#Division
x=22
y=7
#absolute division
z=x/y
print(z)
print(round(z))

#Formatting Methods
print("%f"%(z))
print("%.2f"%(z))
#,always give iine space by default
print("z = {:.3}".format(z))
print("z=",format(z,'.2f'))
```

```python
x=33
y=13
#x=33, y=13, x/y=2.22
print("x = {},y = {},z = {:.3}".format(x,y,x/y) )
```

# Iterators

```python
#Iterator of even number

class even:
    def __init__(self,start,max_limit):
        if start%2==0:
            self.start=start
        else:
            self.start=start+1
        self.max_limit=max_limit
    def __iter__(self):
        self.val=self.start-2
        return self
    def __next__(self):
        if self.val<=self.max_limit :
            self.val+=2
            return self.val
        else:
            raise StopIteration("Max Limit exceeded")
```

```python
one=even(1,1000)
one=iter(one)
```

```python
for i in range(1000):
    print(next(one),end=" ")
```

```python
#Iterator of Prime number

import math
class Prime:
    def __init__(self,start):
        self.start=start
    def __iter__(self):
        self.max_limit=1000
        self.counter=0
        return self
    def prime(self):
        num=self.start
        x=round(math.sqrt(num)+1)
        for var in range(2,x):
            if num % var == 0:
                self.start+=1
                return self.prime()
        else:
            self.start+=1
            return num
    def __next__(self):
        while self.counter <= self.max_limit :
            self.counter+=1
            return self.prime()
        else:
            raise StopIteration
```

```python
one=Prime(2)
one=iter(one)
```

```python
for i in range(30):
    print(next(one),end=" ")
```

```python
#To print prime number in a range without storing prie numbers

from math import sqrt
x = Prime(1)
x = iter(x)
c = 0
for var in x :
    print(var,end=', ')
    c = c+1
print("Total numbers = ",c)
```

```python
#Iterator to generate palindrom number in a range

class Palindrom:
    def __init__(self,start,end):
        self.start=start
        self.end=end
    def __iter__(self):
        return self
    def palindrom(self):
        num=self.start
        x=num
        s=0
        while x:
            r=x%10
            s=s*10+r
            x=x//10
        if s==num:
            self.start+=1
            return num
        else:
            self.start+=1
            return self.palindrom()
    def __next__(self):
        if self.start<=self.end:
            return self.palindrom()
        else:
            raise StopIteration
```

```python
#Creating Object

one=Palindrom(1,2000)
```

```python
#Starting iterator

one=iter(one)
```

```python
#To print values without storing these values

for i in range(1999):
    print(next(one),end=" ")
```

```python
#To print values without storing these values

for i in one:
    print(i,end=" ")
```

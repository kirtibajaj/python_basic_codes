```python
def even(num,max_limit):
    c = 0
    while c <= max_limit :
        if num % 2 == 0 :
            yield num
            num = num + 1
            c = c + 1
        else :
            num = num + 1
            c = c + 1
    
```

```python
x = even(2,1000)
for var in x :
    print(var,end=" ")
```

```python
def prime(num,end):
    while num<=end:
        for i in range(2,round(math.sqrt(num)+1)):
            if num%i==0:
                num+=1
                break
        else:
            yield num
            num+=1
```

```python
import math
x=prime(2,1000)
for i in x:
    print(i,end=" ")
```

```python
def armstrong(start,end):
    while start<=end:
        x=start
        s=0
        while x:
            r=x%10
            s=s*10+r
            x=x//10
        if s==start:
            yield start
            start+=1
        else:
            start+=1
    
```
